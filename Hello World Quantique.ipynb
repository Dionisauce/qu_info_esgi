```python
# utilsons un simulateur quantique
```

```python
# importons Aer =  le module des simulateurs
```

```python
# from qiskit import Aer
```

```python
#on cree notre moteur de circuit
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
#on lance le calcul quantique
```

```python
job = backend.run(circ)
```

```python
#affichons le resulat
```

```python
result = job.result()
```

```python
#on a simplement des vecteurs d'etat
```

```python
outputstate = result.get_statevector(circ, decimals = 3)
```

```python
print(outputstate)
```

```python
#affichons un histogramme
```

```python
from qiskit.visualization import plot_state_city
```

```python
plot_state_city(outputstate)
```
